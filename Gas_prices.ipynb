Here we have a simple database showing how much each country spent on gas (usd per gallon) from 1990-2008.
I researched which country spent the most on gas, who spent the least on gas, and how much did the UK spent on gas in pounds (their local currency).

```python
#import operators
import pandas as pd
import seaborn as sb
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
gas=pd.read_csv( 'Gas_prices.csv')
```

```python
#check data types
gas.dtypes
```

```python
#understand the database
gas.shape
gas.index
gas.head()
```

```python
#screen data 
gas.min()
gas.max()
gas.describe
gas.value_counts()
gas.sum()
```

```python
#visual representation of the database
plt.title('Gas Prices Per Year (USD)')


plt.plot(gas.Year, gas.USA, label='US')
plt.plot(gas.Year, gas.Australia, label='AUS')
plt.plot(gas.Year, gas.Canada, label='Canada')
plt.plot(gas.Year, gas.France, label='France')
plt.plot(gas.Year, gas.Germany, label='GER')
plt.plot(gas.Year, gas.Italy, label='Italy')
plt.plot(gas.Year, gas.Japan, label='Japan')
plt.plot(gas.Year, gas.Mexico, label='MEX')
plt.plot(gas.Year, gas['South Korea'], label='S. Korea')
plt.plot(gas.Year, gas.UK, label='UK')

plt.xticks(gas.Year[::2])
plt.xlabel('Year')
plt.ylabel('USD per gallon')
plt.legend()
plt.figure(figsize=(8,10))



plt.show()
```

```python
#Italy and France spent the most on gas between 1990-2008
plt.subplot(1,2,1)
plt.bar(gas.Year, gas.France, color='#fa2a3f')
plt.title('USD per Gallon spent on Gas',fontdict={'fontweight':'bold'})

plt.subplot(1,2,2)
plt.bar(gas.Year, gas.Italy, color='#8f4d03')
plt.title('USD per Gallon spent on Gas',fontdict={'fontweight':'bold'})
plt.show()
```

```python
#Italy takes the prize!
gas.Italy.sum()
```

```python
gas.France.sum()
```

```python
#which country spent the least on gas from 1990-2008?
plt.scatter(x=gas.Year, y=gas.Mexico, label='Mexico')
plt.scatter(x=gas.Year, y=gas.USA, label='USA')
plt.xlabel('Year')
plt.ylabel('USD per Gallon spent')
plt.xticks(gas.Year[::2])
plt.legend()
plt.show()
```

```python
gas.Mexico.sum()
```

```python
#America wins this round!
gas.USA.sum()
```

```python
gas.USA.plot(kind='box', title='Sum of US Gas Spendings')
plt.show()
```

```python
#The United Kingdom''s spendings on gas from 1990-2008, depicted in pounds-their local currency
UK=gas.UK*0.75
sb.scatterplot(x='Year', y='UK', hue='Year',data=gas, palette='plasma')
plt.show()
```
