Here is a sample database detailing gas prices over time in USD per gallon from 1990-2008 in core first world nations.
The graphs shown detail the rising and declining prices in each country, and the UKs' spendings on gas (petrol) in pounds (their local currency)

```python
import pandas as pd
import seaborn as sb
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
gas=pd.read_csv( 'Gas_prices.csv')
```

```python
#getting to know the data...all functions can be used
gas.shape
gas.index
gas.min()
gas.max()
gas.describe
gas.value_counts()
gas.sum()
gas.dtypes
gas.head(5)
```

```python
gas.index=pd.RangeIndex(start=1, stop=len(gas) + 1)
gas.head(5)
```

```python
plt.title('Gas Prices Per Year (USD)')


plt.plot(gas.Year, gas.USA, label='US')
plt.plot(gas.Year, gas.Australia, label='AUS')
plt.plot(gas.Year, gas.Canada, label='Canada')
plt.plot(gas.Year, gas.France, label='France')
plt.plot(gas.Year, gas.Germany, label='GER')
plt.plot(gas.Year, gas.Italy, label='Italy')
plt.plot(gas.Year, gas.Japan, label='Japan')
plt.plot(gas.Year, gas.Mexico, label='MEX')
plt.plot(gas.Year, gas['South Korea'], label='S. Korea')
plt.plot(gas.Year, gas.UK, label='UK')

plt.xticks(gas.Year[::2])
plt.xlabel('Year')
plt.ylabel('USD per gallon')
plt.legend()
plt.figure(figsize=(8,10))



plt.show()
```

```python
plt.suptitle('Highest Gas Prices From 1990-2008')

plt.subplot(1,2,1)
plt.bar(gas.Year, gas.France, color='#fa2a3f')
plt.title('France')

plt.subplot(1,2,2)
plt.bar(gas.Year, gas.Italy, color='#8f4d03')
plt.title('Italy')

plt.show()
```

```python
#Lowest gas prices overall--least growth
plt.scatter(x=gas.Year, y=gas.Mexico, label='Mexico')
plt.scatter(x=gas.Year, y=gas.USA, label='USA')
plt.xlabel('Year')
plt.ylabel('USD per Gallon')
plt.xticks(gas.Year[::2])
plt.legend()
plt.show()
```
